```python
# Dependencies
from splinter import Browser
from bs4 import BeautifulSoup as soup
import requests
from webdriver_manager.chrome import ChromeDriverManager  #
```

```python
# URL of page to be scraped
url = 'https://newjersey.craigslist.org/search/sss?sort=rel&query=guitar'
url2 = 'https://www.nhl.com/'
```

```python
# Retrieve page with the requests module
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
browser.visit(url2)


```

```python
html = browser.html

html_soup = soup(html, 'html.parser')
```

```python
TextDisplayObject = html_soup.find_all('h5')

print(type(TextDisplayObject))



```

```python
#find find first one
#findall find all
```

```python
showcasevideos = html_soup.find_all('h5', class_='showcase__headline')

headlines = []

for headline in showcasevideos:
    print(headline.text.strip())
    print(type(headline))
    headlines.append(headline.text)

print(headlines)   
    
```

```python
video_urls = html_soup.find_all('article', attrs={'data-video-url' : True}, class_="showcase__list-item")
for item in video_urls:
    print(item['data-video-url'])
    print(type(item))
```

```python
showcasevideos = html_soup.select('article .showcase__meta-data h5')
for headline in showcasevideos:
    print(headline.attrs['class'])
```

```python
video_urls = html_soup.select('article[data-video-url]')
for video in video_urls:
    print(video['class'])
    print(type(video['class']))
```

```python
browser.quit()
```
